```python
## Part 3 : Differentiating Data Extraction and Plotting - Peaking Period

# Create a list to store the Peaking Period in Days from the point of Daily Cases crossing 10
Interval = []

# Extract the Date when Daily Cases Peaked/Recorded Maximum so far for Each Country.
bdp = fullbd['date'][fullbd['new_cases'] == fullbd['new_cases'].max()]
indp = fullindia['date'][fullindia['new_cases'] == fullindia['new_cases'].max()]
pakp = fullpak['date'][fullpak['new_cases'] == fullpak['new_cases'].max()]
itap = fullitaly['date'][fullitaly['new_cases'] == fullitaly['new_cases'].max()] 
spap = fullspain['date'][fullspain['new_cases'] == fullspain['new_cases'].max()]
gerp = fullgermany['date'][fullgermany['new_cases'] == fullgermany['new_cases'].max()]

# Extract the Date when Daily Cases Crossed 10 per Day for Each Country.
bds = fullbd['date'][fullbd['new_cases'] == fullbd['new_cases'][fullbd['new_cases'] > 10].min()]
inds = fullindia['date'][fullindia['new_cases'] == fullindia['new_cases'][fullindia['new_cases'] > 10].min()]
paks = fullpak['date'][fullpak['new_cases'] == fullpak['new_cases'][fullpak['new_cases'] > 10].min()]
itas = fullitaly['date'][fullitaly['new_cases'] == fullitaly['new_cases'][fullitaly['new_cases'] > 10].min()]
spas = fullspain['date'][fullspain['new_cases'] == fullspain['new_cases'][fullspain['new_cases'] > 10].min()]
gers = fullgermany['date'][fullgermany['new_cases'] == fullgermany['new_cases'][fullgermany['new_cases'] > 10].min()]

# Create two Serieses to store the Dates for Peak and Initial(Crossing 10) infection.
Peaktime = pd.Series([bdp.values,indp.values,pakp.values,itap.values,spap.values,gerp.values], index = ['Bangladesh','India','Pakistan','Italy','Spain','Germany'])
Starttime = pd.Series([bds.values,inds.values,paks.values,itas.values,spas.values,gers.values], index = ['Bangladesh','India','Pakistan','Italy','Spain','Germany'])

# Calculate the Difference to Get the Number of Days it took to Reach Peak Infection in Each of the Countries.
for i in country_list : 
    Interval.append((Peaktime[i] - Starttime[i]).astype('timedelta64[D]') / np.timedelta64(1, 'D')) 

country_list2 = ['Bangladesh\n(Still\nPeaking)','India\n(Still\nPeaking)','Pakistan\n(Still\nPeaking)','Italy','Spain','Germany']    

# Plot the Intervals & Save the Figure
plt.subplot(2,2,3)
plt.bar(range(len(Peaktime)),Interval, color = colors,  edgecolor='black')
plt.xticks(range(6),country_list2)
plt.ylabel('Number of Days to Reach COVID-19 Peak')
plt.title('Days taken to Reach COVID-19 Infection Peak/Maximum')
plt.savefig('Cases, Deaths, Peak & Ratio.pdf')
```
